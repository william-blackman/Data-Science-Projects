```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
set(data.continent)
```

```python
data_2007 = data[data.year == 2007]
```

```python
asia_2007 = data_2007[data_2007.continent == 'Asia']
europe_2007 = data_2007[data_2007.continent == 'Europe']
```

```python
asia_2007.head()
```

```python
europe_2007.head()
```

```python
print(len(set(asia_2007.country)))
print(len(set(europe_2007.country)))
```

```python
print('Mean GDP Per Captia in Asia: {:.2f}'.format(asia_2007.gdpPerCapita.mean()))
print('Median GDP Per Captia in Asia: {:.2f}'.format(asia_2007.gdpPerCapita.median()))
print('Mean GDP Per Captia in Europe: {:.2f}'.format(europe_2007.gdpPerCapita.mean()))
print('Median GDP Per Captia in Europe: {:.2f}'.format(europe_2007.gdpPerCapita.median()))
```

```python
plt.subplot(2, 1, 1)
plt.title('Distribution of GDP per Capita')
plt.hist(asia_2007.gdpPerCapita, 20, range=(0, 50000), edgecolor='black')
plt.ylabel('Asia')
plt.subplot(2, 1, 2)
plt.hist(europe_2007.gdpPerCapita, 20, range=(0, 50000), edgecolor = 'black')
plt.ylabel('Europe')
plt.show()
```

```python
data_1997 = data[data.year == 1997]
```

```python
americas_1997 = data_1997[data_1997.continent == 'Americas']
europe_1997 = data_1997[data_1997.continent == 'Europe']
```

```python
plt.subplot(2, 1, 1)
plt.title('Comparison of 1997 Life Expectancy by Continent')
plt.hist(americas_1997.lifeExpectancy, 20, range=(55, 80), edgecolor='black')
plt.ylabel('Americas')
plt.subplot(2, 1, 2)
plt.hist(europe_1997.lifeExpectancy, 20, range=(55, 80), edgecolor='black')
plt.ylabel('Europe')
plt.xlabel('Life Expectancy')
```

```python
US = data[data.country == 'United States']
china = data[data.country == 'China']
```

```python
plt.plot(US.year, US.gdpPerCapita)
plt.xlabel('year')
plt.ylabel('GDP per Capita')
```

```python
plt.plot(china.year, china.gdpPerCapita)
plt.xlabel('year')
plt.ylabel('GDP per Capita')
```

```python
plt.plot(US.year, US.gdpPerCapita)
plt.plot(china.year, china.gdpPerCapita)
```

```python
US.gdpPerCapita.iloc[0]
```

```python
us_growth = US.gdpPerCapita / US.gdpPerCapita.iloc[0] * 100
china_growth = china.gdpPerCapita / china.gdpPerCapita.iloc[0] * 100
```

```python
plt.title('GDP Per Capita Growth (first year = 100)')
plt.plot(US.year, us_growth)
plt.plot(china.year, china_growth)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('GDP per capita growth')
```
