```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
data_1952 = data[data.year == 1952]
asia_1952 = data_1952[data_1952.continent == 'Asia']
africa_1952 = data_1952[data_1952.continent == 'Africa']
```

```python
#making sure the data sets have a relatively similar number of countries
print(len(set(asia_1952.country)))
print(len(set(africa_1952.country)))
```

```python
#comparing the mean & median life expectancies in both continents
print('Mean Life Expectancy in Asia: {:.2f}'.format(asia_1952.lifeExpectancy.mean()))
print('Median Life Expectancy in Asia: {:.2f}'.format(asia_1952.lifeExpectancy.median()))
print('Mean Life Expectancy in Africa: {:.2f}'.format(africa_1952.lifeExpectancy.mean()))
print('Median Life Expectancy in Africa: {:.2f}'.format(africa_1952.lifeExpectancy.median()))
```

```python
#setting a 2 row by 1 column canvas, and working on row 1
plt.subplot(2, 1, 1)
plt.title('Life Expectancy Distribution')
#plotting the life expectancy in asia with 20 bins and a set range
plt.hist(asia_1952.lifeExpectancy, 20, range=(25,65), edgecolor='black')
#setting a y axis range so the charts are comprable
plt.ylim([0, 11])
plt.ylabel('Asia\n # of countries')
#working on row 2
plt.subplot(2, 1, 2)
#plotting the life expectancy in Africa with 20 bins and a set range
plt.hist(africa_1952.lifeExpectancy, 20, range=(25,65), edgecolor='black')
plt.ylabel('Africa\n # of countries')
plt.xlabel('Life Expectancy in years')
plt.ylim([0, 11])
plt.show()
```

```python
#evaluating the time series of Life Expectancy in Asia and Africa
#first assigning all of the data for asia & africa to variables
asia = data[data.continent == 'Asia']
africa = data[data.continent == 'Africa']
```

```python
#setting up the same 2 row 1 column canvas
plt.subplot(211)
plt.title('Life Expectancy Over Time by Continent')
#plotting life expectancy in asia over time
plt.scatter(asia.year, asia.lifeExpectancy, 5, alpha=0.3)
plt.ylim([25,85])
plt.ylabel('Asia')
plt.subplot(212)
#plotting life expectancy in africa over time
plt.scatter(africa.year, africa.lifeExpectancy, 5, alpha=0.3)
plt.ylim([25,85])
plt.ylabel('Africa')
plt.xlabel('Year')
```
