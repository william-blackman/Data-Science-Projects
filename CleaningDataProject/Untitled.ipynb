```python
import pandas as pd
```

```python
#assigning the csv file to the variable data
data = pd.read_csv('Realestate.csv')
```

```python
#showing the first several rows of the data
data.head()
```

```python
#checking the data types of all of the columns
data.dtypes
```

```python
#assigning the house age column to the variable age to make the code cleaner
age = data.houseAge
```

```python
#normalizing the house age and assigning those values to the variable norm
norm = (age - age.mean()) / age.std()
```

```python
#calling norm to visually check the normalized data
norm
```

```python
#normalizing that data to the 0 - 1 range
minmax = (age - age.min()) / (age.max() - age.min())
minmax
```

```python
#creating a new column and assigining the normalized house age data 
data['houseAgeStandardized'] = minmax
```

```python
#dropping a column
data.drop('numberOfConvenienceStores', axis=1, inplace=True)
```

```python
data.head()
```

```python
#renaming the transaction column to transactionDate
data.rename(columns={"transaction": "transactionDate"}, inplace=True)
```
